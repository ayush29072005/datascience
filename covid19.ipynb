```python
import pandas as pd
df=pd.read_csv("covid_19_clean_complete.csv")
print(df.columns)
```

```python
df["CFR(%)"]=(df["Deaths"]/df["Confirmed"])*100
print(df)
print(df.head)
```

```python
df_no_nan=df[df.notnull().all(axis=1)]
print(df_no_nan)
```

```python
df["Recovery_Rate(%)"]=(df["Recovered"]/df["Confirmed"])*100
print(df)
print(df.head)
```

```python
df_no_nan=df[df.notnull().all(axis=1)]
print(df_no_nan)
```

```python
df["Active_Cases"]=df["Confirmed"]-df["Deaths"]+df["Recovered"]
print(df)
print(df.head)
```

```python
df['Cases_per_100k']=(df['Confirmed']/df['Confirmed']+df['Deaths']+df['Recovered'])*100000
print(df)
print(df.head)
```

```python
df_no_nan=df[df.notnull().all(axis=1)]
print(df_no_nan)
```

```python
import numpy as np
df["Log_Confirmed_Cases"]=np.log1p(df["Confirmed"])
print(df)
```

```python
df["Risk_Score"]=(df["CFR(%)"]*0.7)+(df["Cases_per_100k"]*0.3)
print(df)
```
